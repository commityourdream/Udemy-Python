```python
import pandas as pd
```

```python
stats=pd.read_csv('DemographicData.csv')
```

```python
import matplotlib as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 9,4
```

```python
##Distribution
#Seaborn is used to nicely represnt data 
```

```python
vis1=sns.distplot(stats["Internet users"])
```

```python
vis1=sns.distplot(stats["Internet users"],bins=10) #here bins=10 means we are diving total plot in 10 parts
```

# Box_Plots###

```python
vis2=sns.boxplot(data=stats,x='Income Group',y='Birth rate')
```

```python
vis3=sns.lmplot(data=stats,x='Internet users',y='Birth rate')
```

```python
vis3=sns.lmplot(data=stats,x='Internet users',y='Birth rate',fit_reg=False)
```

```python
vis3=sns.lmplot(data=stats,x='Internet users',y='Birth rate',fit_reg=False,hue='Income Group',size=8)
```

# Keywords_Argument(kws)_########

```python
vis3=sns.lmplot(data=stats,x='Internet users',y='Birth rate',fit_reg=False,hue='Income Group',size=5,scatter_kws={"s":100})
```
